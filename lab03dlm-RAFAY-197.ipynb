```python
#ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#01
import pandas as pd
import numpy as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Load data
print("ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#01 ")
data = pd.read_csv("Salary_dataset.csv")
#print(data)
# Extract features and target
X = data[['YearsExperience']]
Y = data['Salary']
# Create and train the model
model = LinearRegression()
model.fit(X, Y)
# Predict salary for 3 and 10 years of experience
experience_years = [[3], [10]]
predicted_salaries = model.predict(experience_years)
print(f"The Predicted Salary for 3 Years of Experience is: {predicted_salaries[0]}")
print(f"The Predicted Salary for 10 Years of Experience is: {predicted_salaries[1]}")
# Plot
plt.scatter(X, Y, color='ORANGE', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.show()
```

```python
#ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#02
import pandas as pd
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
# Load original dataset (without outlier)
print("ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#02")
data_original = pd.read_csv("Salary_dataset.csv")
X_orig = data_original[['YearsExperience']]
Y_orig = data_original['Salary']
# Train model on original data
model_orig = LinearRegression()
model_orig.fit(X_orig, Y_orig)
Y_pred_orig = model_orig.predict(X_orig)
# Load modified dataset (with outlier)
data_outlier = pd.read_csv("Salary_dataset_updated.csv")
X_outlier = data_outlier[['YearsExperience']]
Y_outlier = data_outlier['Salary']
# Train model on data with outlier
model_outlier = LinearRegression()
model_outlier.fit(X_outlier, Y_outlier)
Y_pred_outlier = model_outlier.predict(X_outlier)
# Predict salary for 6 years of experience
pred_6_orig = model_orig.predict([[6]])[0]
pred_6_outlier = model_outlier.predict([[6]])[0]
# Print comparison
print("== ORIGINAL DATASET ==")
print("Slope:", model_orig.coef_[0])
print("Intercept:", model_orig.intercept_)
print("MSE:", mean_squared_error(Y_orig, Y_pred_orig))
print("Predicted Salary for 6 YOE:", pred_6_orig)
print("\n== DATASET WITH OUTLIER ==")
print("Slope:", model_outlier.coef_[0])
print("Intercept:", model_outlier.intercept_)
print("MSE:", mean_squared_error(Y_outlier, Y_pred_outlier))
print("Predicted Salary for 6 YOE:", pred_6_outlier)
# Plotting both regression lines
plt.figure(figsize=(10, 6))
plt.scatter(X_outlier, Y_outlier, color='purple', label='Data with Outlier')
plt.plot(X_outlier, Y_pred_outlier, color='red', label='Regression (With Outlier)')
plt.scatter(X_orig, Y_orig, color='orange', label='Original Data')
plt.plot(X_orig, Y_pred_orig, color='blue', label='Regression (Original)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Comparison of Linear Regression: With vs Without Outlier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#03
import pandas as pd
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
# Load data
print("ABDUL RAFAY / 2022F-BSE-197 / LAB#03 / TASK#03")
data = pd.read_csv("Salary_dataset.csv")
print(data)
# Extract features and target
X = data[['YearsExperience']]  
Y = data['Salary']             
# Train the model
model = LinearRegression()
model.fit(X, Y)
# Find slope and intercept
slope = model.coef_[0]
intercept = model.intercept_
print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
# Predict salaries for training data
Y_pred = model.predict(X)
# Calculate Mean Squared Error
mse = mean_squared_error(Y, Y_pred)
print("Mean Squared Error:", mse)
# Predict salary for 6 years of experience
predicted_salary = model.predict([[6]])
print("The Predicted Salary Is:", predicted_salary[0])
# Plot
plt.scatter(X, Y, color='purple', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.show()
```
